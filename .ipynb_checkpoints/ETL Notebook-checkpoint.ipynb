```python
import pandas as pd
```

```python
#Loading .csv file into dataframe
shark_tank_file = 'shark_tank_data.csv'

load = pd.read_csv(shark_tank_file)
load.head()
```

```python
list(load.columns)
```

```python
#Some light data cleanup
load['ASK_Equity'] = load['ASK_Equity'].replace({'%':''}, regex=True)
load['DEAL_Equity'] = load['DEAL_Equity'].replace({'%':''}, regex=True)
load.head()
```

```python
pitch_file = 'shark_tank_pitch_detail.csv'

load2 = pd.read_csv(pitch_file)
load2.head()
```

```python
#Loading raw dataframe to Postgres DB
from sqlalchemy import create_engine

engine = create_engine('postgresql://postgres:ENTERPASSWORD@localhost:5432/SharkTank')
#ENTERPASSWORD
conn = engine.connect()
```

```python
load.to_sql('raw_numbers', engine)
load2.to_sql('pitch_descriptions', engine)
```
